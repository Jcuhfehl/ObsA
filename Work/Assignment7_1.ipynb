# Interpretation of the CMD

A detailed discussion of what one can learn from CMDs and other types of observations belongs in other courses and the next steps will, in general, depend on the specific scientific question that is being addressed. As an example, we will here outline how the CMD could be used to infer the distance and age of M37.

Here below is a plot of the colour-magnitude diagrams of the Hyades and the Pleiades with absolute $M_V$ magnitude on the vertical axis. We see the main sequences in the two clusters overlap nicely, but that the main sequence turn-off is brighter in the Pleiades. This indicates that the Pleiades are younger than the Hyades (about 110-160 Myr vs. 680 Myr). 

```python
import matplotlib.pyplot as plt
import numpy as np

MVH, BVH = np.loadtxt('Hyades.txt',usecols=(1,2), unpack=True)
MVP, BVP = np.loadtxt('Pleiades.txt',usecols=(1,2), unpack=True)

plt.plot(BVH, MVH, '.', label='Hyades')
plt.plot(BVP, MVP, '.', label='Pleiades')
plt.legend()
plt.xlim(-0.3, 1.8)
plt.ylim(12,-4)
plt.xlabel('B-V')
plt.ylabel(r'M$_V$')
plt.minorticks_on()
plt.show()
```

By plotting M37 on the same diagram, we can get information about:
- The distance (from the difference between the apparent magnitudes $m$ and absolute magnitudes $M$ at a given colour)
- The age (from the locatio of the main sequence turn-off)

A complication is that we haven't measured B-V for stars in M38, but instead V-r'. Therefore, we need a way to convert between the two. In the Observational Astronomy notes, such a relation is derived from a fit to URAT1 data:

$$
(V-r^\prime) = -0.11 +0.48 \times (B-V)
$$

```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
with open("M37.pickle", 'rb') as file: # 
    M37_V_r, M37_V_apparent  = pickle.load(file)

M37_V_B = (np.array(M37_V_r)+0.11)/0.48
M37_V = np.array(M37_V_apparent) - 9.2

MVH, BVH = np.loadtxt('Hyades.txt',usecols=(1,2), unpack=True)
MVP, BVP = np.loadtxt('Pleiades.txt',usecols=(1,2), unpack=True)

plt.plot(BVH, MVH, '.', label='Hyades')
plt.plot(BVP, MVP, '.', label='Pleiades')
plt.plot(M37_V_B, M37_V, '.', label="M37")
plt.legend()
plt.xlim(-0.3, 1.8)
plt.ylim(12,-4)
plt.xlabel('B-V')
plt.ylabel(r'M$_V$')
plt.minorticks_on()
plt.show()
```

Comment on the age and distance of M37

The shift needed to make the main sequences overlap is -9.2, the distance is than caluclated by:
$m-M = 5\log{\frac{D}{10pc}}$

$\frac{m-M}{5}=\log{\frac{D}{10pc}}$

$10^{\frac{m-M}{5}}=\frac{D}{10pc}$

$D = 10*10^{\frac{m-M}{5}}pc$

Thus, $D = 10*10^{\frac{9.2}{5}}pc = 692pc$, this is under half the actual distance to M37, this means that more light was measured than should have been, which is a pretty weird result.

$L \propto M^{3.5}$

$\tau \propto \frac{M}{L}$

$\frac{M/L}{M_\odot/L_\odot}=\frac{\tau}{\tau_\odot}$

$\tau = \tau_\odot \frac{M/M_\odot}{L/L_\odot}$

$\tau = \tau_\odot (L/L_\odot)^{-\frac{5}{7}}$

$\tau = \tau_\odot (10^{\frac{M-M_\odot}{-2.5}})^{-\frac{5}{7}}$

The turnoff point of M37 is at around $M_V=3$, so $M-M_\odot=3-4.83=-1.83$
So the age is
$\tau = 10^9 (10^{\frac{-1.83}{-2.5}})^{-\frac{5}{7}}yr = 3.0*10^8yr$, this is just under the actual estimates of M37's age, so it's a pretty good result.
